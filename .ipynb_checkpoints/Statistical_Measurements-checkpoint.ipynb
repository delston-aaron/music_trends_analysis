```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file name)
df = pd.read_csv('spotify_datasets/top_10000_1950-now.csv')

# View the first few rows of the dataset
df.head()
```

```python
df.describe()
print(df.head())
print(df.shape)
print(df.describe())
print(df.info())
```

```python
print(df.isnull().sum())

```

```python
print(df.duplicated().sum())
```

```python
df['Album Release Date'] = pd.to_datetime(df['Album Release Date'].str[:4], format='%Y', errors='coerce')
```

```python
df['release year'] = df['Album Release Date'].dt.year
```

```python
df.head()
```

```python
mean_danceability = df['Danceability'].mean()
print(f"Mean Danceability: {mean_danceability}")
mean_energy = df['Energy'].mean()
print(f"Mean Energy: {mean_energy}")
```

```python
median_tempo = df['Tempo'].median()
print(f"Median Tempo: {median_tempo}")
```

```python
mode_key = df['Key'].mode()
print(f"Mode of Key: {mode_key.values[0]}")

mode_sig = df['Time Signature'].mode()
print(f"Mode of Time Signature: {mode_sig.values[0]}")
```

```python
std_tempo = df['Tempo'].std()
print(f"Standard Deviation of Tempo: {std_tempo}")
```

```python
from scipy.stats import skew

# Example: Calculate skewness for the 'valence' feature
skew_valence = skew(df['Valence'].dropna())
print(f"Skewness of Valence: {skew_valence}")
```
