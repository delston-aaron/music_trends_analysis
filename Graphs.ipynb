```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file name)
df = pd.read_csv('spotify_datasets/top_10000_1950-now.csv')

# View the first few rows of the dataset
df.head()
```

```python
df.describe()
print(df.head())
print(df.shape)
print(df.describe())
print(df.info())
```

```python
print(df.isnull().sum())

```

```python
print(df.duplicated().sum())
```

```python
df['Album Release Date'] = pd.to_datetime(df['Album Release Date'].str[:4], format='%Y', errors='coerce')
```

```python
df['release year'] = df['Album Release Date'].dt.year
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of relevant numerical columns
relevant_columns = ['Danceability', 'Energy', 'Valence', 'Tempo']  # Replace with your columns

for col in relevant_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col.capitalize()}')
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

relevant_features = ['Popularity','Danceability','Energy','Speechiness','Acousticness','Instrumentalness','Liveness','Valence','Tempo']  # Replace with your features

for feature in relevant_features:
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Album Release Date', y=feature)
    plt.title(f'{feature.capitalize()} Over the Years')
    plt.xlabel('Year')
    plt.ylabel(feature.capitalize())
    plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

relevant_features = ['Danceability','Energy','Speechiness','Acousticness','Instrumentalness','Liveness','Valence']
# Plot each feature
for feature in relevant_features:
    plt.plot(df['Album Release Date'], df[feature], label=feature.capitalize(), alpha=0.7)

# Add title and labels
plt.title('Music Features Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Feature Value', fontsize=14)
plt.legend(title='Features', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
