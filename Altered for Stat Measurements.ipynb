```python
import matplotlib.pyplot as plt

#here there are two types of visualizations for the same analysis

plt.figure(figsize=(10, 6))
plt.plot(decade_correlation['Decade'], decade_correlation['Danceability_Energy_Correlation'], marker='o', linestyle='-', color='b')
plt.title('Correlation Between Danceability and Energy Over Decades')
plt.xlabel('Decade')
plt.ylabel('Correlation')
plt.grid(True)
plt.xticks(decade_correlation['Decade'])
plt.show()


# Group by decade and calculate mean values for danceability and energy
decade_trends = df.groupby('Decade')[['Danceability', 'Energy']].mean().reset_index()

# Plot the trends
plt.figure(figsize=(10, 6))

# Plot Danceability
plt.plot(
    decade_trends['Decade'], 
    decade_trends['Danceability'], 
    marker='o', 
    linestyle='-', 
    color='b', 
    label='Danceability'
)

# Plot Energy
plt.plot(
    decade_trends['Decade'], 
    decade_trends['Energy'], 
    marker='s', 
    linestyle='--', 
    color='r', 
    label='Energy'
)

# Add titles and labels
plt.title('Trends in Danceability and Energy Over Decades')
plt.xlabel('Decade')
plt.ylabel('Average Value')
plt.legend()  # Add legend to distinguish lines
plt.grid(True)
plt.xticks(decade_trends['Decade'])  # Ensure decade labels are shown correctly

# Show the plot
plt.show()
```
