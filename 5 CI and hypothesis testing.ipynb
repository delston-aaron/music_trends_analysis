```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file name)
df = pd.read_csv('spotify_datasets/top_10000_1950-now.csv')

# View the first few rows of the dataset
df.head()


```

```python
df.describe()
print(df.head())
print(df.shape)
print(df.describe())
print(df.info())
```

```python
print(df.isnull().sum())

```

```python
print(df.duplicated().sum())
```

```python
df['Album Release Date'] = pd.to_datetime(df['Album Release Date'], errors='coerce')

# Extract only the year from the datetime object
df['Year'] = df['Album Release Date'].dt.year

# Drop rows where 'Year' is NaN (invalid or missing)
df = df.dropna(subset=['Year'])

# Ensure 'Year' is an integer
df['Year'] = df['Year'].astype(int)

df = df[df['Year'] >= 1960]

# Create a 'Decade' column (e.g., 1991 → 1990)
df['Decade'] = (df['Year'] // 10) * 10
```

```python
df.head()
```

```python
df = df.drop(columns=(['Track URI', 'Artist URI(s)',
       'Album URI', 'Album Artist URI(s)',
       'Album Artist Name(s)', 'Album Image URL',
       'Disc Number', 'Track Number', 'Artist Genres','Track Preview URL',
       'ISRC', 'Added By', 'Added At','Album Genres',
       'Label', 'Copyrights']))
```

```python
print(df.columns)
```

```python
import scipy.stats as stats
import numpy as np

# Function to calculate confidence intervals
def calculate_confidence_interval(data, confidence=0.95):
    n = len(data)  # Sample size
    mean = np.mean(data)  # Sample mean
    sem = stats.sem(data)  # Standard error of the mean
    margin = sem * stats.t.ppf((1 + confidence) / 2., n-1)  # Margin of error
    return mean - margin, mean + margin  # Lower and upper bound of CI

# Group by decade and calculate CI for danceability
decade_stats = df.groupby('Decade')['Danceability'].apply(
    lambda x: calculate_confidence_interval(x)
).reset_index(name='Danceability_CI')

# Add mean values for plotting
decade_stats['Danceability_Mean'] = df.groupby('Decade')['Danceability'].mean().values

# Print result
print(decade_stats)
```

```python
# Plot mean and confidence intervals
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.errorbar(decade_stats['Decade'], decade_stats['Danceability_Mean'], 
             yerr=(decade_stats['Danceability_Mean'] - decade_stats['Danceability_CI'].apply(lambda x: x[0])), 
             fmt='o-', label='Danceability CI', color='b')
plt.title('Danceability Trends with Confidence Intervals')
plt.xlabel('Decade')
plt.ylabel('Danceability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filter data for two decades
data_1990s = df[df['Decade'] == 1990]['Danceability']
data_2000s = df[df['Decade'] == 2000]['Danceability']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(data_1990s, data_2000s, nan_policy='omit')

# Output results
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in danceability between 1990s and 2000s.")
else:
    print("Fail to reject the null hypothesis: No significant difference in danceability.")
```
