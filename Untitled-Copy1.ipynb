```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

relevant_features = ['Danceability','Energy','Speechiness','Acousticness','Instrumentalness','Liveness','Valence']
# Normalize features to a 0-1 range
scaler = MinMaxScaler()
normalized_data = df.copy()
normalized_data[relevant_features] = scaler.fit_transform(df[relevant_features])

# Plot normalized features
plt.figure(figsize=(12, 8))
for feature in relevant_features:
    plt.plot(normalized_data['Album Release Date'], normalized_data[feature], label=feature.capitalize(), alpha=0.7)

plt.title('Normalized Music Features Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Normalized Feature Value', fontsize=14)
plt.legend(title='Features', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
